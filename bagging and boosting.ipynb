```python
import numpy as np 
import pandas as pd
```

```python
data=pd.read_table('http://archive.ics.uci.edu/ml/machine-learning-databases/00198/Faults.NNA', sep='\t', header=None)
data.head()
```

```python
data.shape
```

```python
data_rand=data.sample(frac=1.0)
```

```python
dvs=data_rand.iloc[:,27:]
dvs
```

```python
class_labels=dvs.idxmax(axis=1)
```

```python
class_labels
```

```python
IVs=data_rand.iloc[:,:27]
IVs.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,BaggingClassifier, RandomForestClassifier
from sklearn.metrics import cohen_kappa_score, make_scorer
```

```python
x_train, x_test, y_train, y_test= train_test_split(IVs,class_labels,test_size=0.2, random_state=123)
```

```python
def kappa(y,y_pred, **kargs):
    return cohen_kappa_score(y,y_pred)

Kappa= make_scorer(kappa)
```

```python
tree=DecisionTreeClassifier()
rf=RandomForestClassifier()
bagg=BaggingClassifier(base_estimator=tree)
boost=AdaBoostClassifier(base_estimator=tree)
```

```python
param_grid_tree={'criterion':['gini','entropy'],'max_depth':[3,5,7,9]}
param_grid_rf={'n_estimators':[100,200,500],'max_depth':[3,5,7,9]}
param_grid_bagg={'n_estimators':[100,200,500]}
param_grid_boost={'n_estimators':[100,200,500],'learning_rate':[0.0001,0.0003,0.0005]}
```

```python
best_model_tree=GridSearchCV(estimator=tree,param_grid=param_grid_tree,scoring=Kappa,cv=5)
best_model_rf=GridSearchCV(estimator=rf,param_grid=param_grid_rf,scoring=Kappa,cv=5)
best_model_bagg=GridSearchCV(estimator=bagg,param_grid=param_grid_bagg,scoring=Kappa,cv=5)
best_model_boost=GridSearchCV(estimator=boost,param_grid=param_grid_boost,scoring=Kappa,cv=5)
```

```python
best_model_tree.fit(X=x_train,y=y_train)
```

```python
best_model_tree.best_params_
```

```python
cross_val_score(best_model_tree.best_estimator_,X=x_train,y=y_train,cv=5)
```

```python
best_model_rf.fit(X=x_train,y=y_train)
```

```python
best_model_rf.best_params_
```

```python
cross_val_score(best_model_rf.best_estimator_,X=x_train,y=y_train,cv=5)
```

```python
best_model_bagg.fit(X=x_train,y=y_train)
```

```python
best_model_bagg.best_params_
```

```python
cross_val_score(best_model_bagg.best_estimator_,X=x_train,y=y_train,cv=5)
```

```python
best_model_boost.fit(X=x_train,y=y_train)
```

```python
best_model_boost.best_params_
```

```python
cross_val_score(best_model_boost.best_estimator_,X=x_train,y=y_train,cv=5)
```
